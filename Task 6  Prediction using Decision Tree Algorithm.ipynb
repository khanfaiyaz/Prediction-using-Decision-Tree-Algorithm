Create the Decision Tree classifier and visualize it graphically.●The purpose is if we feed any new data to this classifier, it would be able topredict the right class accordingly.●Dataset : https://bit.ly/3kXTdox

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Iris.csv")
data.head()
X=data.drop(["Id","Species"],axis=1)
y=np.array(data["Species"])
features=X.columns
```

```python
features=features.tolist()
```

```python
labels=data["Species"].unique()
labels=labels.tolist()
```

# Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python

y=y.reshape(-1,1)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=140)
```

```python
X_train.shape
```

# Training using Decision TreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
ds=DecisionTreeClassifier(random_state=141)
ds.fit(X_train,y_train)
```

# Model Prediction

```python
ds.predict(X_test)
```

```python
print("The accuracy of given model is",ds.score(X_train,y_train)*100,"%")
```

# Visualization of Descision Tree Classifier

```python
from sklearn import tree
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(ds,
               feature_names = features, 
               class_names=labels,
               filled = True);
fig.savefig('graphimage.png')
```

You can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.
